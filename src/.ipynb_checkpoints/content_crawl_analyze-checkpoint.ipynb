```python
def get_update(Q_dict,Q_df,svg_pkl,svg_file ) :

    with open(Q_dict, 'rb') as f:
        q_dict = pickle.load(f)
        
    with open(Q_df,'rb') as g:
        q_df = pickle.load(g)
    
    
    title = q_dict['title']
    date = q_dict['date']
    
    for idx, title in enumerate(title) :
        q_df.loc[q_df['title'] == title, 'date'] = date[idx]
        
    
    with open(svg_pkl, 'wb') as f :
        pickle.dump(q_df, f)
    
    q_df.to_csv(svg_file, index = False)   
    
    return print("Success")
```

```python
# function for changing type
def datetype(x) :
    #print(row)
    month_dict = {'January' :1, 'February' :2, 'March' :3, 'April':4 , 'May':5,  'June':6, 
                  'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}  
    month = x.split(',')[0].split(' ')
    year =x.split(',')[1]
    month.append(year)

    d = datetime.datetime(int(month[2]), month_dict[month[0]], int(month[1]))
    d = d.strftime("%Y-%m-%d")
    
    return d
```

```python
def slice_list(dates, dur) :
    '''
    dates = list of date
    dur = duration i.g) by week, two weeks, one month
    '''
    dur = 14
    N = int(np.floor(len(dates)/dur))
    

    for i in range(N) :
        slice_list = []
        if i < N-1 :
            i = i*dur
            slice_list = dates[i:i+dur]
        else :
            i = i*dur
            slice_list = dates[i:]
        
        i = int(i/dur)
        #print(slice_list)
        pick_name = "list"+str(i)+"th_by" +str(dur)
        
        with open("pkl/slice_pkl/two_week/"+pick_name+".pkl", "wb") as f :
            pickle.dump(slice_list, f)
    
    return slice_list 
    
```

```python
class VaccineAnalyze() :
    
    def __init__(self, df_sorted, dates, dur) :
        
        self.df = df_sorted
        self.dates = dates
        self.dur  = dur
        self.N = int(np.floor(len(dates)/dur))
    
    '''
    Generating dataframe
    '''
    
    def slice_list(self) :
        '''
        dates = list of date
        dur = duration i.g) by week, two weeks, one month
        '''

        for i in range(self.N) :
            slice_list = []
            if i < N-1 :
                i = i*self.dur
                slice_list = self.dates[i:i+self.dur]
                print(slice_list)
            else :
                i = i*self.dur
                slice_list = self.dates[i:]

            i = int(i/dur)
            pick_name = "list"+str(i)+"th_by" +str(self.dur)
            with open("pkl/slice_pkl/two_week/"+pick_name+".pkl", "wb") as f :
                pickle.dump(slice_list, f)

        return 0
    
    def slice_df(self) :

        for i in range(self.N) :

            with open("pkl/slice_pkl/two_week/list"+str(i)+"th_by"+str(self.dur)+".pkl", "rb") as f : 
                sliced_list= pickle.load(f)

            sliced_df = self.df[self.df['date'].isin(sliced_list)]
            pick_name= "df"+str(i)+"th_by" +str(self.dur)
            with open("pkl/slice_pkl/two_week/"+pick_name+".pkl", "wb") as f:
                pickle.dump(sliced_df,f)
                
        return 0
    
    '''
    Analyze Data
    '''    
    
    def basic_stats(self) :
        
        n_question = {}
        n_answers = {}
        n_tags = {}
        for i in range(self.N) :
            
            pick_name= "df"+str(i)+"th_by" +str(self.dur)
            with open("pkl/slice_pkl/two_week/"+pick_name+".pkl", "rb") as f :
                sliced_df = pickle.load(f)
            
            days = sorted(list(set(sliced_df['date'])))
            #print(days)
            duration = str(days[0])+'~'+str(days[-1])
            n_question[duration] = len(sliced_df)
            n_answers[duration] = [np.median(list(sliced_df['N_answer'])),np.mean(list(sliced_df['N_answer']))]
            
            tag_list = []
            for tags in sliced_df['tags'] :
                tag_list.extend(tags)    
            n_tags[duration] = tag_list
            
            '''
            tag_name = 'tag' +str(i) +'th_by'+str(self.dur)
            with open("pkl/slice_pkl/two_week/"+tag_name+".pkl", "wb") as f :
                pickle.dump(tag_list, f)
            '''
        
        return n_question, n_answers, n_tags
        
        
```
